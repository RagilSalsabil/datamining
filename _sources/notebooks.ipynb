READ FILE

```python
df = pd.read_csv('https://raw.githubusercontent.com/RagilSalsabil/datamining/main/Survei_Medsos.csv')
```

DATA FRAME

```python
df.head()
```

DATA SAMPLE

```python
data_sample = df[
    [
      "Jenis_Kelamin",
      "Pekerjaan",
      "Usaha_Melepaskan_Diri",
      "Butuh_Aplikasi",
      "Pernah_Memakai_Aplikasi_Pengaturan_Waktu"
    ]
  ].replace(["Laki-laki", "Pernah", "Ya"],1).replace(["Perempuan", "Tidak"], 0).head().reset_index(drop=True)
data_sample
```

FORMULA FOR SYMMETRIC BINARY VARIABLES

```python
def d(i,j):
  r = 0
  q = 0
  s = 0
  t = 0
  
  for col in ["Jenis_Kelamin","Usaha_Melepaskan_Diri","Butuh_Aplikasi","Pernah_Memakai_Aplikasi_Pengaturan_Waktu"]:
    x1 = data_sample[col][i]
    x2 = data_sample[col][j]

    if x1 == 1 and x2 == 1: q+=1
    elif x1 == 1 and x2 == 0: r+=1
    elif x1 == 0 and x2 == 1: s+=1
    else: t+=1

  return (r+s)/(q+r+s+t)

```

```python
print(d(0,1))
print(d(0,2))
print(d(0,3))
```
